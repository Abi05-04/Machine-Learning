```python
###
###STEP 1:LOAD &CLEAN DATA
import pandas as pd
df=pd.read_csv("Retail sales.csv")
df.drop_duplicates()
df.fillna(0)
df["Order Date"]=pd.to_datetime(df["Order Date"])
df
```

```python
###STEP:2 FEATURE ENGINEERING:DATA EXTRACTION
df["Year"]=df["Order Date"].dt.year
df["Month"]=df["Order Date"].dt.month
df["Day"]=df["Order Date"].dt.day
df["Weekday"]=df["Order Date"].dt.weekday
df
```

```python
#STEP3: CATEGORICAL ENCODING
#ENCODE CATEGORY&REGION
df=pd.read_csv("Retail sales.csv")
df_encoded=pd.get_dummies(df,columns=["Product","Region"],drop_first=True)
df_encoded
```

```python
##STEP:4 BUSINESS FEATURE CREATION(REAL-WORLD)
### 1.PROFIT MARGIN

df_encoded["Profit Margin"]=df_encoded["Profit"] / df_encoded["Sales"]
df_encoded["Profit Margin"].fillna(0)
df_encoded
```

```python
##STEP:4 BUSINESS FEATURE CREATION(REAL-WORLD)
####2.HIFH-PROFIT LABEL(CLASSIFICATION TARGET)
df_encoded["High_Profit"]=df_encoded["Profit"].apply(lambda x:1 if x>df_encoded["Profit"].median() else 0)
df_encoded
```

```python
###### STEP:5 FEATURE SCALING(ML ESSENTIAL)

from sklearn.preprocessing import StandardScaler
#df=pd.read_csv(r"C:\Users\DELL\Downloads\Retail_sales.csv")
scaler=StandardScaler()
num_cols=["Quantity","Sales","Profit","Profit Margin"]
df_encoded[num_cols]=scaler.fit_transform(df_encoded[num_cols])
df_encoded
```

```python
###### STEP 6:FEATURE & TARGRT SELECTION
### REGRESSION
X=df_encoded.drop(["Sales","Order Date"],axis=1)
Y=df_encoded["Sales"]
df_encoded
```
